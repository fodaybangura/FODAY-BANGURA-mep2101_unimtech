# [Problem 1]Creating Dummy Data

```python
import numpy as np
#create  the variables
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
seed = np.random.seed(0)
rand_num = np.random.multivariate_normal(mean, cov, 500)
print(rand_num)
rand_num.shape
```

# [Problem 2] Visualization by scatter plot

```python

from matplotlib import pyplot as plt
%matplotlib inline
plt.title("Visualization by scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(rand_num[:,0], rand_num[:,1])
```


# [Problem 3] Visualization by histogram

```python

import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('x1')
plt.ylabel('frequency')
plt.title('Histogram of x1')
plt.hist(rand_num[:,0])
plt.xlim(left=-5)
plt.show()
```

```python

import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('x2')
plt.ylabel('frequency')
plt.title('Histogram of x2')
plt.hist(rand_num[:,1])
plt.xlim(right=2)
plt.show()
```

# [Problem 4] Addition of data


```python

mean = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]
seed = np.random.seed(0)
rand_num1 = np.random.multivariate_normal(mean, cov, 500)
print(rand_num1)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.title("Two in one scatter plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(rand_num[:,0], rand_num[:,1], label='0', marker='o')
plt.scatter(rand_num1[:,0], rand_num1[:,1], label='1', marker='o')
plt.legend()
plt.show()
```

# [Problem 5] Data combination

```python

#Using concatenate
data_combine = np.concatenate((rand_num, rand_num1),axis=0)
print(data_combine)
data_combine.shape
```

```python
## Using vstack
combined_data = np.vstack((rand_num, rand_num1))
combined_data.shape
```
